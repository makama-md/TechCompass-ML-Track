### Handling Missing Values

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv(r"C:\Users\makama\Pictures\TechCompass ML Track\data\fistula_data.csv")
```

```python
# identify missing values/NaN
missing_values = dataset.isna().sum()  # or dataset.isnull()
print(missing_values)
```

```python
# Check for blank strings across the entire DataFrame
df_is_blank = dataset.applymap(lambda x: isinstance(x, str) and x.strip() == '').sum()
print(df_is_blank)
```

```python
# Replace empty strings with NaN

dataset = dataset.replace("", np.nan, inplace=True)
```

```python
print(dataset)
```

```python
# Select the relevant features within the dataset

data_selected = dataset[['Age', 'Age_marriage', 'Age_fistula_months', 'Parity',
                         'Marital_status', 'Urban_Rural', 'Education', 'Occupation',
                         'Booking_status', 'Place_delivery', 'Mode_delivery',
                         'Weight', 'Height_cm', 'Urethra',
                         'Vaginal_fibrosis', 'Diagnosis', 'Duration_admission',
                         'Outcome_discharge', 'Labour_duration']]
```

```python
# Replace missing values for specific columns
data_selected['Age_marriage'] = data_selected['Age_marriage'].fillna(data_selected['Age_marriage'].median())
data_selected['Occupation'] = data_selected['Occupation'].fillna(data_selected['Occupation'].median())
data_selected['Urethra'] = data_selected['Urethra'].fillna(data_selected['Urethra'].median())
data_selected['Labour_duration'] = data_selected['Labour_duration'].fillna(data_selected['Labour_duration'].median())
```
